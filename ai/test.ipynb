```python

def longest_word(word_list):
    longest = ""
    for word in word_list:
        if len(word) > len(longest):
            longest = word
    return longest  
print(longest_word(["apple" , "banana" , "cherry" , "watermelon"]))

    
```

```python
def index_longest(word_list):
    longest = ""
    longest_index = 0
    for word in word_list:
        if len(word) > len(longest):
            longest_index = longest_index+1
            longest = word   
    return longest , longest_index

word,longest_index = index_longest(["apple" , "banana" , "cherry" , "watermelon"])

print(word , longest_index)
```

```python
list_numbers = [1,3,4,5,6,7,2]

def prime_number(list_numbers):
    primes = []
    for num in list_numbers:
        if num > 1:
            for i in range(2, num):
                if (num % i) == 0:
                    break
            else:
                primes.append(num)
    return primes
print(prime_number(list_numbers))



```

```python
class product:
    def __init__(self, name="", price=100, quantity=10):
        self.name = name
        self.price = price
        self.quantity = quantity

    def sell(self):
        if self.quantity == 0:
            print("not found")
        else:
            self.quantity -= 1

    def restock(self, quantity):
        self.quantity += quantity
```

```python
p = product("apple", 10, 3)

p.sell()
print(p.quantity)

p.restock(5)
print(p.quantity)
```

```python
class calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b == 0:
            return "Error: Division by zero"
        return a / b
    
calc = calculator()
print(calc.add(10,5))
print(calc.subtract(10,5))
print(calc.multiply(10,5))
print(calc.divide(10,5))
print(calc.divide(10,0))    
```

```python
class bank :
    def __init__(self , name ,account_name , balance , address):
        self.name = name
        self.account_name = account_name
        self.balance = balance
        self.address = address
    def deposit(self , amount):
        self.balance += amount
        print(f"Deposited: {amount}, New Balance: {self.balance}")
    def withdraw(self , amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"Withdrew: {amount}, New Balance: {self.balance}")

account = bank("ayman" , "ayman123" , 1000 , "street")
account.deposit(500)
account.withdraw(2000)
account.withdraw(300)

        
```

```python
class Student:
    def __init__(self, name, age, courses):
        self.name = name
        self.age = age
        self.courses = courses

    def enroll(self, course):
        self.courses.append(course)

    def get_info(self):
        return f"Name: {self.name}, Age: {self.age}, Courses: {', '.join(self.courses)}"
    

student1 = Student("ayman", 22, ["AI", "Embedded Systems"])
student1.enroll("Data Science")
print(student1.get_info())  
```

```python
class book :
    def __init__(self , titel , author):
        self.titel = titel
        self.author = author

class library :
    def __init__(self):
        self.books = []
    
    def add_book(self , book):
        self.books.append(book)
    
    def remove_book(self , book):
        self.books.remove(book)
    
    def list_books(self):
        for book in self.books:
            print(f"Title: {book.titel} , Author: {book.author}")

    
lib = library()
book1 = book("Python" , "ahmed")
book2 = book("Data Science" , "mohamed")
lib.add_book(book1)
lib.add_book(book2)
lib.list_books()
lib.remove_book(book1)
lib.list_books()
      
```
